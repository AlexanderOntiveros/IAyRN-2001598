```python
#4 SelecciÃ³n y entrenamiento del modelo
print(df.info())
print(df.describe())
```

## Generar VisualizaciÃ³n de Datos


```python
#5. Prueba del modelo.
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

target_labels = {0: 'Target 0', 1: 'Target 1'}
colors = {0: 'blue', 1: 'red'}

for target_value in df['target'].unique():
    subset = df[df['target'] == target_value]
    plt.scatter(
        subset.index,
        subset['distancia_cm'],
        label=target_labels[target_value],
        color=colors[target_value],
        alpha=0.6
    )

plt.xlabel('Ãndice del Dato')
plt.ylabel('Distancia (cm)')
plt.title('DistribuciÃ³n de Distancia por Target')
plt.legend(title='Target')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
